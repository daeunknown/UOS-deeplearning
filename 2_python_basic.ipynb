```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.xlabel("x value")
plt.ylabel("y value")

plt.title("sin/cos")

plt.plot(x, y_sin, label = 'sin')
plt.plot(x, y_cos, label = 'cos', linestyle = 'dashed')
plt.legend()

plt.show()
```

```python
#예제 1

def factorial(n):
    output = 1
    for i in range(1, n+1):
        output *= i
    return output

print(factorial(3))
```

```python
#예제 3 - 재귀함수

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

factorial(3)
```

```python
students = [
    {"name" : "김다은", "korean" : 89, "math" : 90, "english" : 71, "science" : 78},
    {"name" : "김나은", "korean" : 69, "math" : 96, "english" : 77, "science" : 89},
    {"name" : "김가은", "korean" : 80, "math" : 90, "english" : 100, "science" : 81},
    {"name" : "김라은", "korean" : 95, "math" : 88, "english" : 80, "science" : 94}
]

print("이름", "총점", "평균", sep = "\t")

for student in students:
    score_sum = student["korean"] + student["math"] + student["english"] + student["science"]
    score_average = score_sum / 4
    
    print(student["name"], score_sum, score_average, sep = "\t")
```

```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius
    def get_circumference(self):
        return 2 * math.pi * self.radius
    def get_area(self):
        return math.pi * (self.radius ** 2)
    
circle = Circle(10)
print("둘레:", circle.get_circumference())
print("넓이:", circle.get_area())
```
