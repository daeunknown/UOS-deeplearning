```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#노드에서 이뤄지는 계산 실습

x = np.array([1, 2, 3])
w = np.array([[1, 3, 4],
             [4, 5, 2],
             [1, 1, 1]])
b = np.array([6, 4, 2])

u = np.dot(x, w) + b
print(u)
```

```python
#계단 함수(step function) 구현

def step_function(x):
    return np.where(x<=0, 0, 1)

x = np.linspace(-5, 5)
y = step_function(x)

plt.plot(x, y)
plt.show()
```

```python
#시그모이드 함수(Sigmoid function) 구현

def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))

x1 = np.linspace(-5, 5)
y = sigmoid_function(x1)

plt.plot(x1, y)
plt.show()
```

```python
#tanh 함수 구현

def tanh_function(x):
    return np.tanh(x)

#def tanh_function(x):
#    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

y = tanh_function(x)

plt.plot(x, y)
plt.show()
```

```python
#ReLU 함수 구현

def relu_function(x):
    return np.where(x<=0, 0, x)

y = relu_function(x)

plt.plot(x, y)
plt.show()
```

```python
#Leaky ReLU 구현

def leaky_relu_function(x):
    return np.where(x<=0, 0.01*x, x)

y = leaky_relu_function(x)

plt.plot(x, y)
plt.show()
```

```python
#Leaky ReLU 출력값 확인 -> 엄청 작아

x2 = np.linspace(-5, -3)
y = leaky_relu_function(x2)
print(y)
```

```python
#항등 함수 구현

y = x

plt.plot(x, y)
plt.show()
```

```python
#Softmax 구현

def softmax_function(x):
    return np.exp(x) / np.sum(np.exp(x))

#실행
y = softmax_function(np.array([1, 2, 3]))
print(y)   #원-핫 인코딩 전
print(np.sum(y))
```
